```python
import os
import pandas as pd
import numpy as np
movies = pd.read_csv('./data/input/IMDB-Movie-Data.csv')
```

```python
def bins_movies(x):
    if x < 1000:
        return 1
    elif x > 1000 and x < 10000:
        return 2
    elif x > 10000 and x < 100000:
        return 3
    elif x > 100000 and x < 1000000:
        return 4
    else:
        return 5
    
    
```

```python
movies['bins'] = movies['Votes'].apply(bins_movies)
```

```python
movies['revenue_per_minute']= movies.apply(lambda x: x['Revenue (Millions)']/ x['Runtime (Minutes)'], axis = 1)
```

```python
thriller = movies['Genre'].str.find('Thriller', start=0, end=None)
comedy = movies['Genre'].str.find('Comedy', start=0, end=None)
```

```python
thriller.sum()
```

```python
comedy.sum()
```

```python
def genre_mod(x):
    if thriller:
        return x+1
    elif comedy:
        return x-1
    else:
        return x
        
```

```python
movies['New Ranking'] = movies['Rating'].apply(genre_mod)
```
